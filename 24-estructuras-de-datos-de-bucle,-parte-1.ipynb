## Bucle sobre el diccionario

En Python 3, necesita el método **items()** para recorrer un diccionario:

**world = { "afghanistan":30.55,   
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;          "albania":2.77,  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;          "algeria":39.21 }**

**for key, value in world.items() :
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    print(key + " -- " + str(value))**
    
¿Recuerda el diccionario *europe* que contenía los nombres de algunos países europeos como clave y sus capitales como valor correspondiente? ¡Continúa y escribe un bucle para iterar sobre él!

**Actividad**

* Escriba un bucle *for* que pase por cada clave: par de valores de *europe*. En cada iteración debe imprimirse "*the capital of x is y*" , donde **x** es la clave e **y** es el valor de la tupla.

```python
# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin',
          'norway':'oslo', 'italy':'rome', 'poland':'warsaw', 'austria':'vienna' }
          
# Iterate over europe
for key, value in europe.items():
    print("the capital of " + key + " is " + value)
```

## Recorrer la matriz de Numpy

Si se trata de una matriz 1D Numpy, recorrer todos los elementos puede ser tan simple como:

**for x in my_array :  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    ...**
    
Si se trata de una matriz 2D Numpy, es más complicado. Una matriz 2D está formada por múltiples matrices 1D. Para iterar explícitamente sobre todos los elementos separados de una matriz multidimensional, necesitará esta sintaxis:

**for x in np.nditer(my_array) :  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;    ...**
    
Dos conjuntos de Numpy que puede reconocer en el curso de introducción están disponibles en su sesión de Python: *np_heightun* conjunto de Numpy que contiene las alturas de los jugadores de Major League Baseball y *np_baseballun* conjunto de Numpy 2D que contiene las alturas (primera columna) y los pesos (segundo columna) de esos jugadores.

**Actividad**

* Importe el paquete *numpy* con el alias *np*.
* Escriba un bucle *for* que repita todos los elementos *np_height* e imprima "*x inches*"para cada elemento, donde **x** es el valor en la matriz.
* Escriba un bucle *for* que recorra cada elemento de la *np_baseball* matriz y lo imprima.


```python
import pandas as pd
df = pd.read_csv("data/mlb_weight_ht.csv")
baseball = df[['Height(inches)', 'Weight(pounds)']].values.tolist()

# Import numpy as np
import numpy as np

# Create a 2D numpy array from baseball: np_baseball
np_baseball = np.array(baseball)
np_height = np.array(np_baseball[:,0])

# For loop over np_height
for a in np_height:
    print(str(a) + " inches")

# For loop over np_baseball
for b in np.nditer(np_baseball):
    print(b)
```
