## Plot lineal (1)

Con matplotlib, puede crear un montón de tramas diferentes en Python. El diagrama más básico es el diagrama lineal. Aquí se da una receta general.

**import matplotlib.pyplot as plt  
plt.plot(x,y)  
plt.show()**

En el video, ya viste cuánto ha crecido la población mundial en los últimos años. ¿Continuará haciéndolo? El banco mundial tiene estimaciones de la población mundial para los años 1960 hasta 2050. Los años se cargan en su espacio de trabajo como una lista llamada year, y las poblaciones correspondientes como una lista llamada pop.

* **print()** el último elemento de *year* la lista *pop* y de la lista para ver cuál es la población prevista para el año 2050. Usa dos funciones **print()**.
* Antes de comenzar, debe importar *matplotlib.pyplotcomo plt*. *pyplot* es un subpaquete de *matplotlib*, de ahí el punto.
* Use **plt.plot()** para construir un diagrama lineal. yeardebe mapearse en el eje horizontal, popen el eje vertical. No olvide terminar con la función **show()** para mostrar realmente la trama.


```python
import pandas as pd
df = pd.read_csv("data/world_population.csv", sep =';')
year = list(df['year'])
pop = list(df['pop'])

# Print the last item from year and pop
print(year[-1])
print(pop[-1])


# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Make a line plot: year on the x-axis, pop on the y-axis
plt.plot(year,pop)

# Display the plot with plt.show()
plt.show()
```

## Plot lineal (3)

Ahora que ha construido su diagrama de primera línea, comencemos a trabajar en los datos del worldbank. Hay dos listas disponibles para usted:

* *life_exp* que contiene la esperanza de vida de cada país y
* *gdp_cap*, que contiene el PIB per cápita (es decir, por persona) para cada país expresado en miles de dólares estadounidenses.

El PIB representa Producto Interno Bruto. Básicamente representa el tamaño de la economía de un país. Divida esto entre la población y obtendrá el PIB per cápita.

*matplotlib.pyplot* ya está importado como *plt*, por lo que puede comenzar de inmediato.

* Imprima el último elemento tanto de la lista *gdp_cap* como de la lista *life_exp*; Es información mundial.
* Construya un gráfico de líneas, con *gdp_cap* en el eje x y en el eje y *life_exp*. ¿Tiene sentido trazar estos datos en un diagrama lineal?
* No olvide terminar con un comando *plt.show()*, para mostrar realmente la trama.


```python
import pandas as pd

df = pd.read_csv("data/world_bank_data.csv", sep =';', skiprows=1, 
                 names=["Country_Name",
                        "Country_Code",
                        "Continent_Name",
                        "Continent_Code",
                        "Population_1960",
                        "Population_2018",
                        "GDP_per_capita_1960",
                        "GDP_per_capita_2018",
                        "Life_expectancy_at_birth_1960",
                        "Life_expectancy_at_birth_2018"
])
gdp_cap = list(df["GDP_per_capita_2018"][(df["Population_2018"]>0) & (df["GDP_per_capita_2018"]>0) & (df["Life_expectancy_at_birth_2018"]>0)])
life_exp = list(df["Life_expectancy_at_birth_2018"][(df["Population_2018"]>0) & (df["GDP_per_capita_2018"]>0) & (df["Life_expectancy_at_birth_2018"]>0)])
life_exp_1960 = list(df["Life_expectancy_at_birth_1960"][(df["Life_expectancy_at_birth_1960"]>0) & (df["Life_expectancy_at_birth_2018"]>0)])
pop = list(df["Population_2018"][(df["Population_2018"]>0) & (df["GDP_per_capita_2018"]>0) & (df["Life_expectancy_at_birth_2018"]>0)]/1000000)
Continent_Name = list(df["Continent_Name"][(df["Population_2018"]>0) & (df["GDP_per_capita_2018"]>0) & (df["Life_expectancy_at_birth_2018"]>0)])
```

```python
# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

# Print the last item of gdp_cap and life_exp
print(gdp_cap[-1])
print(life_exp)

# Make a line plot, gdp_cap on the x-axis, life_exp on the y-axis
plt.plot(gdp_cap,life_exp)

# Display the plot
plt.show()
```

## Diagrama de dispersión (1)

Cuando tiene una escala de tiempo a lo largo del eje horizontal, el diagrama lineal es su amigo. Pero en muchos otros casos, cuando intenta evaluar si existe una correlación entre dos variables, por ejemplo, el diagrama de dispersión es la mejor opción. A continuación se muestra un ejemplo de cómo construir un diagrama de dispersión.

**import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()**

Continuemos con el gráfico gdp_cap versus life_exp, el PIB y los datos de esperanza de vida para diferentes países en 2016. ¿Quizás un gráfico de dispersión será una mejor alternativa?

Nuevamente, el paquete *matplotlib.pyplot* está disponible como *plt*.

**Actividad**

* Cambie el diagrama lineal que está codificado en el guión a un diagrama de dispersión.
* Una correlación quedará clara cuando visualice el PIB per cápita en una escala logarítmica. Agrega la línea *plt.xscale('log')*.
* Termine su secuencia de comandos con plt.show()para mostrar la trama.


```python
# Change the line plot below to a scatter plot
plt.scatter(gdp_cap, life_exp)

# Put the x-axis on a logarithmic scale
plt.xscale('log')

# Show plot
plt.show()
```
